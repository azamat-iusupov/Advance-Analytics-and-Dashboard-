### The below ananlysis shows how important to know the differences betweeen Cost and Unit Cost and how they impact to Revenue

```python
# let see the correlation between Cost and Revenue for the further analysis
sns.lmplot( y = 'Cost',x = 'Revenue', data = subset2)
plt.title('Correlation_btw_Revenue_and_Cost')
```

### As we can see in the map above, we have a positive correlation between Cost and Revenue. In many businesses, revenue is generated through the sale of products or services. The cost of these products or services (cost of goods sold, COGS) directly affects the revenue. Higher costs typically mean higher prices charged to customers, leading to higher revenue.

```python
# Let us see what is the correlation between Unit Cost and the Revenue

sns.lmplot(y = 'Unit_Cost', x = 'Revenue', data = subset2)
plt.title('Corr_btw_Unit_Cost_and_Revenue')
```

### What we see here, as the Unit Cost reaches 2500(units/euro/usd) it stays unchangable, but the revenue is going up. We do not know, if the company do not whant to spend more money to the Unit Cost and improve the products or they have no more production capacities. We have also two dots (maybe outliers, may not) which broght a lot of Revenue to the Company.

### Why the Distinction  between Cost and Unit Cost Matters
### Pricing: Knowing the unit cost helps in setting the selling price to ensure profitability.
### Cost Control: Understanding the total cost helps in budgeting and financial planning, while the unit cost helps in identifying cost-saving opportunities at the unit level.
### Decision Making: Helps management make informed decisions about production levels, scaling operations, and optimizing resource allocation.By differentiating between total cost and unit cost, businesses can better manage their finances, set appropriate prices, and maintain profitability.

# 5 Categorical plot. I took Customer Age to do the analysis in Categorical Plot.

```python
# Let's see histogram plot by Customer Age
sns.histplot(df['Customer_Age'], bins = 30, kde = True)
plt.title('Histogram of Customer_Age with KDE')
```

### Let's create a 4 (four) Categories of Age

```python
# Creating a Categories
df.loc[df['Customer_Age'] <= 17, 'Age_category'] = 'Young_17'
df.loc[(df['Customer_Age'] >= 18) & (df['Customer_Age'] <= 41), 'Age_category'] = 'Middle_Age_18-41'
df.loc[(df['Customer_Age'] >= 42) & (df['Customer_Age'] <= 59), 'Age_category'] = 'Upper_Middle_Age_42-59'
df.loc[df['Customer_Age'] >= 60, 'Age_category'] = 'Senior_over_60'
```

```python
# Checking the Categories
df['Age_category'].value_counts(dropna = False)
```

```python
# Creating the Plot 'Revenue by Age Categories'
sns.set(style="ticks")
g = sns.catplot(x='Customer_Age', y='Revenue', hue='Age_category', data=df)
plt.title('Revenue_by_Age_category')
plt.xticks(ticks=[], labels=[]) # to delete numbers in x axis, because they overlap each others
```

```python
# Saving the vizualisation
g.figure.savefig(os.path.join(path, 'Analysis', 'Revenue_by_Age_category.png'))
```

### So, what can we see in the above categorical plot: 
### The Young Group represented by small group of people. Group has relatively small impact to the Revenue. 
### The Middle Age Group has the most Impact and high density. It brings the most of Revenue. 
### Upper_Middle_Age Group a little bit less, but still brings the Revenue but with less density of people age, as the Middle_Age Group. 
### Senior Group has less density and Revenue. The tendency has to become drastically low.
