```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load data into a pandas DataFrame
df = pd.read_csv(r"C:\Users\azikb\Downloads\Bike Sales in Europe and USA\Original Data\archive\Sales.csv")
```

```python
# Droping duplicates
df_dups = df[df.duplicated()]
df_dups
# Let's unlimit the display
pd.set_option('display.max_columns', None)
```

```python
df = df.drop_duplicates()
df
```

```python
# Checking data for non a number
df_nan = df.isnull().sum()
```

```python
df_nan
```

```python
# Checking data for mix-type
for col in df.columns.tolist():
  weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```

#### Checking data for outliers

```python
# First let's see what columns we have.
print(df.columns)
```

```python
columns = ['Order_Quantity', 'Unit_Cost', 'Unit_Price','Profit', 'Cost', 'Revenue']

# Create a boxplot
boxplot = sns.boxplot(data=df[columns])
plt.title('Outliers in Variables')
boxplot.set_xlabel('Variables')
boxplot.set_ylabel('Counts')
plt.show()
```
